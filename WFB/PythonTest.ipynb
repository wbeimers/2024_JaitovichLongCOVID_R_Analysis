```python
subset = df.loc[(df['padj'] < 0.05) & (df['NES'] > 1)]
subset = subset.sort_values("padj")
subset = subset.iloc[0:10]
subset['ome'] = ['protein'] * 5 + ['transcript'] * 5
subset["ypos"] = range(len(subset))

fig, ax = plt.subplots(dpi=300, figsize=(3, 2))

ax.barh(y=subset['ypos'], width=subset['NES'])

ax.set_ylim(bottom=len(subset), top=0)
```

```python
sns.barplot(data=subset,
           x="NES",
           y="name",
           hue='ome')
```
